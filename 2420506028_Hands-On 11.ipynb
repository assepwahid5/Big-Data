```python
from pymongo import MongoClient, InsertOne, UpdateOne, DeleteOne

# Establish client connection
client = MongoClient('mongodb://localhost:27017/')
db = client['university_db']
courses_collection = db['courses']

# Bulk insert of courses with student enrollments
operation = [
    UpdateOne({'course': 'Math 101'},{'$set': {'enrollments': 200, 'department': 'Mathematics'}},upsert=True),
    DeleteOne({'course': 'CS 102', 'enrollments': 25, 'department': 'Computer Science'}),
    DeleteOne({'course': 'History 201', 'enrollments': 20, 'department': 'History'}),
    UpdateOne({'course': 'Physics 202'}, {'$set':{'enrollments': 152, 'department': 'Physics'}},upsert=True)
]
courses_collection.bulk_write(operation)
print('Courses inserted successfully.')

for course1 in db.courses.find():
    print(f"Course: {course1['course']}, Enrollments: {course1['enrollments']}, Department: {course1['department']}")
```

```python
# Average enrollment per department using aggregation
pipeline = [
    {'$group': {'_id': '$department', 'average_enrollment': {'$avg': '$enrollments'}}}
]
for result in courses_collection.aggregate(pipeline):
    print(result)

print("")

for course1 in db.courses.aggregate(pipeline):
    print(f"Departemen: {course1['_id']}, Rata-rata penerimaan: {course1['average_enrollment']}")
```

```python
# Maximum enrollment per department
pipeline = [
    {'$group': {'_id': '$department', 'max_enrollment': {'$max': '$enrollments'}}}
]
for course2 in db.courses.aggregate(pipeline):
    print(f"Departemen: {course2['_id']}, Maksimal penerimaan: {course2['max_enrollment']}")
```

### 4. Data Transformation using $project and $addFields
- **Exercise 4**: Use `$project` to rename and only show fields: `course_name`, `department_name`, and `enrollments`.
- **Task 4**: Use `$addFields` to create a new field `enrollment_category` where enrollments > 20 are 'high' and others 'low'.


```python
# Projection to rename fields
pipeline = [
    {'$project': {'course_name': '$course', 'department_name': '$department', 'enrollments': 1}}
]
for result in courses_collection.aggregate(pipeline):
    print(result)

print("")
print("")
print("")

for course2 in db.courses.aggregate(pipeline):
    print(f"Course_Name: {course2['course_name']}, department_name: {course2['department_name']}, Enrollment: {course2['enrollments']}")
```

```python
# Adding enrollment category field based on enrollments
pipeline = [
    {'$addFields': {'enrollment_category': {'$cond': {'if': {'$gt': ['$enrollments', 20]}, 'then': 'high', 'else': 'low'}}}}
]
for result in courses_collection.aggregate(pipeline):
    print(result)

for course2 in db.courses.aggregate(pipeline):
    print(f"Course_Name: {course2['course_name']}, department_name: {course2['department_name']}, Enrollment_Category: {course2['enrollment_category']}")
```

### Homework for Students
- **Homework 1**: Perform an aggregation to get a count of courses per department.
- **Homework 2**: Use `$match` and `$group` together to filter and get only courses with enrollments over 25 in 'Computer Science'.
- **Homework 3**: Apply `$lookup` to join `courses` collection with `students` collection based on student enrollments.


```python
from pymongo import MongoClient

client = MongoClient('mongodb://localhost:27017/')
db = client['university_db']
courses = db['courses']
pipeline1 = [
    {'$group': {'_id': '$department', 'total_courses': {'$sum': 1}}}
]
for result in courses.aggregate(pipeline1):
    print(result)
```

```python
from pymongo import MongoClient

client = MongoClient('mongodb://localhost:27017/')
db = client['university_db']
courses = db['courses']

pipeline2 = [
    {'$match': {'department': 'Computer Science', 'enrollments': {'$gt': 25}}},
    {'$group': {'_id': '$course', 'total_enrollments': {'$sum': '$enrollments'}}}
]

for result in courses.aggregate(pipeline2):
    print(result)
```

```python
from pymongo import MongoClient
client = MongoClient('mongodb://localhost:27017/')
db = client['university_db']
courses = db['courses']
pipeline3 = [
    {'$lookup': {
        'from': 'students',
        'localField': 'enrollments',
        'foreignField': 'student_enrollments',
        'as': 'student_details'
    }}
]
for result in courses.aggregate(pipeline3):
    print(result)
```
