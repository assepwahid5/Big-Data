```python
# Import library
from pymongo import MongoClient

# Koneksi ke MongoDB lokal
client = MongoClient("mongodb://localhost:27017/")

# Pilih database dan koleksi
db = client['company_db']
collection = db['employees']
print("Koneksi ke MongoDB berhasil!")

# --- INSERT DATA ---

# Contoh 1: Insert satu data
employee_data = {
    'name': 'Alice',
    'department': 'Finance',
    'age': 29,
    'salary': 4500
}
collection.insert_one(employee_data)
print("Data berhasil dimasukkan (1 data)")

# Contoh 2: Insert beberapa data sekaligus
multiple_employees = [
    {'name': 'Bob', 'department': 'IT', 'age': 31, 'salary': 5200},
    {'name': 'Charlie', 'department': 'HR', 'age': 26, 'salary': 4100},
    {'name': 'Diana', 'department': 'IT', 'age': 34, 'salary': 6000}
]
collection.insert_many(multiple_employees)
print("banyak data berhasil dimasukan")

# --- READ DATA (QUERY) ---
print("\nData karyawan saat ini:")
for employee in collection.find():
    print(employee)

# --- UPDATE DATA ---
# Update data karyawan bernama Bob
collection.update_one(
    {'name': 'Bob'},                # kondisi pencarian
    {'$set': {'salary': 5500}}      # perubahan yang dilakukan
)
print("\nData Bob telah diperbarui")

# Tampilkan hasil update
for employee in collection.find({'name': 'Bob'}):
    print(employee)

# --- DELETE DATA ---
# Hapus data karyawan bernama Charlie
collection.delete_one({'name': 'Charlie'})
print("\nData Charlie telah dihapus")

# Tampilkan hasil akhir koleksi
print("\nData akhir setelah operasi CRUD:")
for employee in collection.find():
    print(employee)
```

```python
from pymongo import MongoClient

# --- Koneksi ke MongoDB ---
client = MongoClient('mongodb://localhost:27017/')
db = client['company_db']
collection = db['employees']

# --- CREATE (Insert Data) ---
# Data pertama (sudah ada di contoh)
employee_data = {
    'name': 'Alice',
    'department': 'Finance',
    'age': 29,
    'salary': 4500
}
collection.insert_one(employee_data)
print('Data Alice inserted')

# Tambahkan beberapa data tambahan
employees = [
    {'name': 'Bob', 'department': 'IT', 'age': 32, 'salary': 6000},
    {'name': 'Charlie', 'department': 'HR', 'age': 27, 'salary': 4000},
    {'name': 'Diana', 'department': 'Marketing', 'age': 35, 'salary': 5500}
]
collection.insert_many(employees)
print('Beberapa data tambahan inserted')

# --- READ (Query Data) ---
print('\nData setelah insert:')
for employee in collection.find():
    print(employee)

# --- UPDATE (Ubah Data) ---
# Contoh: update salary Bob jadi 6500
collection.update_one({'name': 'Bob'}, {'$set': {'salary': 6500}})
print('\nData Bob berhasil diupdate')

# --- DELETE (Hapus Data) ---
# Contoh: hapus data Charlie
collection.delete_one({'name': 'Charlie'})
print('Data Charlie berhasil dihapus')

# --- Tampilkan hasil akhir ---
print('\nData akhir dalam koleksi:')
for employee in collection.find():
    print(employee)
```

```python
from pymongo import MongoClient

# --- Koneksi ke MongoDB ---
client = MongoClient('mongodb://localhost:27017/')
db = client['company_db']
collection = db['employees']

# --- Data dulu, baru kolom ---
data = [
    ("Asep", "M", 23, 2700, "IT"),
    ("Rina", "F", 28, 4100, "HR"),
    ("Budi", "M", 34, 3000, "IT"),
    ("Dina", "F", 26, 3800, "Finance"),
    ("Tono", "M", 30, 3200, "Finance")
]

columns = ["Name", "Gender", "Age", "Salary", "Department"]

# --- Konversi data ke dictionary agar bisa disimpan ke MongoDB ---
documents = [dict(zip(columns, row)) for row in data]

# --- CREATE (Insert Data) ---
collection.insert_many(documents)
print(" Data berhasil dimasukkan ke MongoDB!\n")

# --- READ (Tampilkan Semua Data) ---
print("Data saat ini:")
for emp in collection.find({}, {"_id": 0}):
    print(emp)

# --- UPDATE (Ubah Data) ---
collection.update_one({"Name": "Asep"}, {"$set": {"Salary": 3000}})
print("\nGaji Asep berhasil diupdate menjadi 3000.\n")

# --- DELETE (Hapus Data) ---
collection.delete_one({"Name": "Tono"})
print("Data Tono berhasil dihapus.\n")

# --- READ LAGI (Lihat Hasil Akhir) ---
print("Data setelah update dan delete:")
for emp in collection.find({}, {"_id": 0}):
    print(emp)
```

```python
# Query Aggregation untuk mencari rata-rata gaji per departemen
pipeline = [
    {'$group': {'_id': '$department', 'average_salary': {'$avg': '$salary'}}}
]
for result in collection.aggregate(pipeline):
    print(result)
```

```python
from pymongo import MongoClient

client = MongoClient("mongodb://localhost:27017/")
db = client["tokoOnline"]

data = [
    ["P001", "Laptop ASUS Vivobook", "Elektronik", 9500000, 15],
    ["P002", "Mouse Logitech M170", "Aksesoris", 150000, 40],
    ["P003", "Keyboard Rexus MX9", "Aksesoris", 350000, 25],
    ["P004", "Headset JBL Tune", "Audio", 720000, 30],
    ["P005", "Monitor Samsung 24\"", "Elektronik", 1850000, 12],
    ["P006", "Flashdisk Sandisk 32GB", "Penyimpanan", 75000, 50],
    ["P007", "Harddisk WD 1TB", "Penyimpanan", 650000, 20],
    ["P008", "Charger Anker 25W", "Elektronik", 280000, 35],
    ["P009", "Powerbank Xiaomi 10000mAh", "Elektronik", 300000, 28],
    ["P010", "Kabel Type-C Ugreen", "Aksesoris", 80000, 60]
]

columns = ["product_id", "product_name", "category", "price", "stock"]

# Ubah list data menjadi list of dict (berdasarkan urutan kolom)
documents = [dict(zip(columns, row)) for row in data]

db.products.insert_many(documents)
print('data telah dimasukan')
```

```python
avg_price = list(db.products.aggregate([
    {"$group": {"_id": None, "avgPrice": {"$avg": "$price"}}}
]))[0]["avgPrice"]

# Cari produk dengan harga di atas rata-rata
above_avg = db.products.find({"price": {"$gt": avg_price}})

print(f"Rata-rata harga: {avg_price}")
for product in above_avg:
    print(product)
```
